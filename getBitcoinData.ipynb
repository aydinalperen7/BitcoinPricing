```python
import time
import random
from pytrends.request import TrendReq
## PYTRENDS DOES NOT WORK

def make_request_with_exponential_backoff(api_call):
    retries = 5  # Maximum number of retries
    backoff_factor = 1  # Initial delay factor

    for i in range(retries):
        try:
            response = api_call()  # Your API call function
            if response.status_code == 200:
                return response
        except TooManyRequestsError:
            sleep_time = backoff_factor * (2 ** i) + random.uniform(0, 1)  # Exponential backoff
            time.sleep(sleep_time)
        except Exception as e:
            raise e  # For any other exceptions, raise them
    raise TooManyRequestsError("Too many failed attempts")

# Usage
# You would replace 'your_api_call_function' with the actual function that performs the API request
pytrends = TrendReq()
keyword = 'bitcoin'
pytrends.build_payload(kw_list=[keyword], timeframe='today 1-d')


response = make_request_with_exponential_backoff(pytrends.interest_over_time())
```
